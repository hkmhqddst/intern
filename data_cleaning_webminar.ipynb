```python
import re
```

```python
def filter_telephone(phone_number):

    try :
        if phone_number[-2] == '.':
            phone_number = phone_number[:-2]
            
        temp = re.findall(r'\d+', phone_number)
        filt_telephone = "".join(temp)
    
    except:
        return 'not valid'
    
    try :
        init_val = filt_telephone.index('8')
    except : 
        init_val = 99
    
    if init_val != 99:
        temp = '62' + filt_telephone[init_val:]
        if len(temp) in range(10,15):
            return temp
        else :
            return 'not valid'
    else :
        return 'not valid'
```

```python
df['phone_number'] = df.apply(lambda x: filter_telephone(x['phone_number']), axis = 1)
```

```python
def filter_channel (webinar_marketing_channel):
        
        list_channel = ['Instagram', 'Google', 'Facebook', 'Telegram', 'Email', 'Whatsapp group', 'Whatsapp Blast']
        
        if webinar_marketing_channel.lower() in ['wa','whatsApp','whatsapp chat','wesan whatsapp','ada yg ngirim ke wa ','whatsapp personal','broadcast whatsapp', 'whatsapp']:
            rep = 'Whatsapp blast'
            
        elif webinar_marketing_channel == 'Whatsapp Grup' : 
            rep = 'Whatsapp group'
            
        elif webinar_marketing_channel == 'Fb' :
            rep = 'Facebook'
        
        elif webinar_marketing_channel in list_channel :
            rep = webinar_marketing_channel
        
        else :
            rep = 'Others'
            
        
        return rep
```

```python
df['webinar_marketing_channel'] = df.apply(lambda x: filter_channel(x['webinar_marketing_channel']), axis = 1)
```
