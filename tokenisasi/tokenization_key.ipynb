```python
import pandas as pd
```

```python
df = pd.read_csv('D:/intern/My_Free_WordStream_Keywords_1.csv') 
```

```python
import re

def tokenize(txt) :
  tokens = re.split('\W+', txt)
  return tokens

df['keyword_clean_tokenized'] = df['Keyword'].apply(lambda x: tokenize(x.lower()))
```

```python
df.head(50)
```

```python
token_list_raw = list(df.keyword_clean_tokenized)
```

```python
flatlist=[]
for sublist in token_list_raw:
    for element in sublist:
        flatlist.append(element)
print(flatlist)
```

```python
set(flatlist)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
cv_matrix = cv.fit_transform(df['Keyword'])
```

```python
cv.vocabulary_
```

```python
count_array = cv_matrix.toarray()
df = pd.DataFrame(data=count_array,columns = cv.get_feature_names())
```
